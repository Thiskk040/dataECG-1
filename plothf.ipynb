```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("ECG_0000_Man.csv")
print(df)
xAxis  = df["time"]
yAxis = df["HRV_LFHF"]
```

```python
plt.title("LFHF Ratio_0000_Man")
plt.xlabel("Time(Sec)")
plt.ylabel("LFHF Ratio(mW)")
plt.text(150, 0.07, "Using Sampling Step:100", fontsize=6)
plt.grid(visible=True, which='major', axis='both')
plt.plot(xAxis,yAxis)
plt.show()
```
